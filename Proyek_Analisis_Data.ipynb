## Visualization & Explanatory Analysis

### Pertanyaan 1: Apa lima kota dengan jumlah pengguna terbanyak?

Menghitung jumlah pengguna per kota

```python
user_count_per_city = customers_df['customer_city'].value_counts().reset_index()
user_count_per_city = pd.DataFrame(user_count_per_city)
user_count_per_city = user_count_per_city.rename(columns={'customer_city': 'user_count', 'index': 'customer_city'})
```

Mnegambil 5 kota dengan pengguna terbanyak

```python
top_cities = user_count_per_city.nlargest(5, 'user_count')
```

Membuat pivot table dari tabel diatas

```python
pivot_table_result = customers_df.pivot_table(values='customer_id', index='customer_city', aggfunc='count').reset_index()
```

Mengubah nama kolom

```python
pivot_table_result = pivot_table_result.rename(columns={'customer_id': 'user_count', 'customer_city': 'city'})
```

Menampilkan pivot table

```python
pivot_table_result.head()
```

Membuat Visualisasi

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='customer_city', y='user_count', data=top_cities, palette='viridis')
plt.title('Top 5 Cities')
plt.xlabel('City')
plt.ylabel('User Count')
plt.show()
```

### Pertanyaan 2: Berapa persentase peningkatan jumlah pesanan 3 bulan terakhir?

Menghitung tanggal 3 bulan yang lalu dari tanggal maksimum

```python
from datetime import datetime, timedelta

end_date = max(orders_df['order_purchase_timestamp'])
start_date = end_date - timedelta(days=90)
```

Menggunakan data hanya untuk 3 bulan terakhir

```python
filtered_orders_last_3_months = orders_df[(orders_df['order_purchase_timestamp'] >= start_date) & (orders_df['order_purchase_timestamp'] <= end_date)]
```

Menghitung total pesanan selama 3 bulan terakhir dan 3 bulan sebelumnya

```python
total_orders_last_3_months = len(filtered_orders_last_3_months)
total_orders_previous_3_months = len(orders_df[(orders_df['order_purchase_timestamp'] < start_date)])
```

Menghitung persentase

```python
percentage_increase = ((total_orders_last_3_months - total_orders_previous_3_months) / total_orders_previous_3_months) * 100
```

Menampilkan visualisasi

```python
plt.figure(figsize=(10, 6))
plt.plot(['Previous 3 Months', 'Last 3 Months'], [total_orders_previous_3_months, total_orders_last_3_months], marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.title('Order Quantity Over Time')
plt.xlabel('Time Period')
plt.ylabel('Order Quantity')
plt.grid(True)
plt.show()

print("Persentase Peningkatan:", percentage_increase)
```
